> **📥 GİRDİ:** y_test, y_pred  
> **📤 ÇIKTI:** Detaylı rapor, metrik görsellerrme  
### 🔧 TEMEL METRİKLER
- **Precision:** TP/(TP+FP) | **Recall:** TP/(TP+FN)
- **F1:** Harmonik ortalama | **Macro:** Eşit ağırlık | **Weighted:** Örnek sayısına göre

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
print('=' * 50 + '\n📋 CLASSIFICATION REPORT BAŞLATILIYOR\n' + '=' * 50)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train.values.ravel())
y_pred = model.predict(X_test)

print('\n📊 SINIFLANDIRMA RAPORU:')
print('-' * 60)
print(classification_report(y_test, y_pred))
report_dict = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report_dict).transpose()
print(f'🎯 Genel Doğruluk: %{accuracy_score(y_test, y_pred)*100:.2f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
class_rows = df_report.drop(['accuracy','macro avg','weighted avg'], errors='ignore')
metrics_df = class_rows[['precision','recall','f1-score']].astype(float)
sns.heatmap(metrics_df, annot=True, fmt='.3f', cmap='YlGn', vmin=0, vmax=1, ax=axes[0])
axes[0].set_title('Sınıf Bazı Metrikler', fontweight='bold')
x = np.arange(len(metrics_df)); w = 0.25
axes[1].bar(x-w, metrics_df['precision'], w, label='Precision', color='#2196F3')
axes[1].bar(x,   metrics_df['recall'],    w, label='Recall',    color='#FF5722')
axes[1].bar(x+w, metrics_df['f1-score'],  w, label='F1-Score',  color='#4CAF50')
axes[1].set_xticks(x); axes[1].set_xticklabels(metrics_df.index)
axes[1].set_ylim(0, 1.1); axes[1].legend()
axes[1].set_title('Precision/Recall/F1', fontweight='bold')
plt.tight_layout(); plt.show()
print('\n✅ İşlem Tamamlandı.')
```
