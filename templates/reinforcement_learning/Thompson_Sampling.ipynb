# 🎲 THOMPSON SAMPLING (THOMPSON ÖRNEKLEMESİ)

> **🎯 AMAÇ:** Olasılıksal bir yaklaşım (Beta Dağılımı) ile en yüksek ödülü getiren seçeneği bulmak.  
> **📥 GİRDİ:** Ortam Verisi (Tıklama Durumları - 1 veya 0)  
> **📤 ÇIKTI:** Toplam Ödül, Seçilen İlanların Dağılımı  

### ⏱️ NE ZAMAN KULLANILIR?
- UCB'ye göre daha "akıllı" ve dinamik bir keşfetme mekanizması istendiğinde.
- Gerçek zamanlı reklam gösterimi sistemlerinde.
- Dönüşüm oranlarının (Conversion Rate) takibi gereken durumlarda.

### ⚠️ UYARI
Tıpkı UCB'de olduğu gibi burada da **`df` değişkeni**, ortamın kendisini simüle eder (Ground Truth).
Model bir eylem yapar, ortamdan ödül (1) veya ceza (0) alır.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("=" * 40)
print("🎲 THOMPSON SAMPLING BAŞLATILIYOR")
print("=" * 40)
```

```python
# 1. ORTAMIN VE DEĞİŞKENLERİN HAZIRLANMASI
# ---------------------------------------------------
if 'df' not in locals():
    print("⚠️ UYARI: 'df' değişkeni bulunamadı! Lütfen veriyi yükleyin.")
    print("Örnek: df = pd.read_csv('Ads_CTR_Optimisation.csv')")
else:
    print(f"✅ Veri Hazır. Toplam Satır: {len(df)}")

# <--- LÜTFEN BURAYI DÜZENLEYİN
N = 1000 # Toplam işlem sayısı
d = 10   # Toplam seçenek (ilan) sayısı

# Thompson Sampling Değişkenleri
# Beta dağılımı için 'Başarı' (1) ve 'Başarısızlık' (0) sayılarını tutmamız lazım.
birler = [0] * d    # Ödül (1) sayısı
sifirlar = [0] * d  # Ödül alamama (0) sayısı

secilenler = []     # Hangi turda hangi ilanı seçtik?
toplam = 0          # Toplam skor
```

```python
# 2. ALGORİTMA DÖNGÜSÜ
# ---------------------------------------------------
print("\n🚀 Simülasyon Başlıyor... (Olasılıksal Seçim)")

for n in range(0, N):
    ad = 0
    max_th = 0
    
    for i in range(0, d):
        # Beta dağılımından rastgele bir değer çek
        # betavariate(alpha, beta) -> alpha = birler + 1, beta = sifirlar + 1
        rand_beta = random.betavariate(birler[i] + 1, sifirlar[i] + 1)
        
        if rand_beta > max_th:
            max_th = rand_beta
            ad = i

    # Seçilen ilanı kaydet
    secilenler.append(ad)
    
    # --- ÖDÜL MEKANİZMASI (ENVIRONMENT) ---
    odul = df.values[n, ad] # 1 veya 0
    
    if odul == 1:
        birler[ad] += 1
    else:
        sifirlar[ad] += 1
    
    toplam += odul

print(f"✅ Simülasyon Tamamlandı.")
print(f"💰 Toplam Ödül: {toplam}")
```

```python
# 3. SONUÇLARIN GÖRSELLEŞTİRİLMESİ
# ---------------------------------------------------
plt.figure(figsize=(10, 6))
plt.hist(secilenler)
plt.title('Sonuçlar (Thompson Sampling)')
plt.xlabel('İlanlar')
plt.ylabel('Seçilme Sayısı')
plt.grid(True, alpha=0.3)
plt.show()

# En çok seçilen ilanı bul
from collections import Counter
en_cok_secilen = Counter(secilenler).most_common(1)[0]
print(f"🏆 En Başarılı İlan: {en_cok_secilen[0]} (Seçilme Sayısı: {en_cok_secilen[1]})")
```
