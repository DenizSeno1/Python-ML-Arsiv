### 🔧 YÖNTEM SEÇİMİ
| Yöntem | Formül | Ne zaman? |
|--------|--------|-----------|
| StandardScaler | (x-μ)/σ | Normal dağılım |
| MinMaxScaler | (x-min)/(max-min) | 0-1 aralığı |
| RobustScaler | (x-median)/IQR | Outlier çok |
### ⚠️ KRİTİK KURAL
SADECE X_train'de `fit_transform`, X_test'e sadece `transform`!
