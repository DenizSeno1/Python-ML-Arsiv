````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

print("=" * 40)
print("🌀 K-MEANS CLUSTERING BAŞLATILIYOR")
print("=" * 40)

# 1. ELBOW YÖNTEMİ (OPTİMUM KÜME SAYISI)
# ---------------------------------------------------
print("\n🔄 Optimum küme sayısı aranıyor (WCSS)...")

wcss = []
k_range = range(1, 11)

# X_train -> Veri_On_Isleme.py'den gelir. (Sadece bağımsız değişkenler)
for i in k_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(k_range, wcss, marker='o', linestyle='--', color='cyan')
plt.title('Elbow Yöntemi (Kırılma Noktasını Seç)')
plt.xlabel('Küme Sayısı (K)')
plt.ylabel('WCSS Degeri')
plt.grid(True, alpha=0.3)
plt.show()
print("👀 Grafikteki 'dirsek' noktasına bakıp aşağıya K değerini gir.")
````

````python
# 2. MODEL EĞİTİMİ VE GÖRSELLEŞTİRME
# ---------------------------------------------------
# <--- DAYI BURAYI DOLDUR (!!!) Grafiğe bak, karar ver
optimum_k = 3 

print(f"\n🚀 Model {optimum_k} küme ile eğitiliyor...")
kmeans_final = KMeans(n_clusters=optimum_k, init='k-means++', random_state=42)
y_kmeans = kmeans_final.fit_predict(X_train)

print("✅ Kümeleme tamamlandı.")
print("📍 Küme Merkezleri:\n", kmeans_final.cluster_centers_)

# 3. GÖRSELLEŞTİRME (2D)
# ---------------------------------------------------
if hasattr(X_train, 'shape') and X_train.shape[1] >= 2:
    plt.figure(figsize=(12, 7))
    # Her kümeyi farklı renkte çiz
    for i in range(optimum_k):
        plt.scatter(X_train[y_kmeans == i, 0], X_train[y_kmeans == i, 1], 
                    s=100, label=f'Küme {i+1}')

    # Merkezleri çiz
    plt.scatter(kmeans_final.cluster_centers_[:, 0], kmeans_final.cluster_centers_[:, 1], 
                s=300, c='yellow', label='Merkezler', edgecolors='black')
    
    plt.title(f'K-Means Sonuçları (K={optimum_k})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("⚠️ Veri 2 boyuttan az, görselleştirme sadece ilk 2 boyuta göre yapıldı.")
````
