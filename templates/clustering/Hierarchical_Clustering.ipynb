````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

print("=" * 40)
print("🧬 HİYERARŞİK KÜMELEME BAŞLATILIYOR")
print("=" * 40)
````

````python
# 1. DENDROGRAM (KÜME SAYISINA KARAR VERME)
# ---------------------------------------------------
# 'ward' metodu varyansı minimize eder (K-Means gibi)
print("\n🌳 Dendrogram çiziliyor... (Bu işlem biraz sürebilir)")

plt.figure(figsize=(12, 6))
dendrogram = sch.dendrogram(sch.linkage(X_train, method='ward'))

plt.title('Dendrogram')
plt.xlabel('Veri Noktaları (Müşteriler)')
plt.ylabel('Öklid Mesafesi (Uzaklık)')
plt.show()

print("👀 Grafikte en uzun dikey çizgiye bakarak 'optimum K' sayısına karar verin.")
print("   (Yatay bir çizgi kestiğinizde en çok boşluk kalan dikey hat)")
````

````python
# 2. MODEL EĞİTİMİ VE TAHMİN
# ---------------------------------------------------
# <--- LÜTFEN BURAYI DÜZENLEYİN: Grafiğe bakarak K değerini seçin
optimum_k = 3 

print(f"\n🚀 Model {optimum_k} küme ile eğitiliyor...")

# Agglomerative Clustering (Birleştirici Kümeleme)
# -------------------------------------------------------------------------
# PARAMETRELER:
# 1. metric (Mesafe Metriği) - (Eski sürümlerde 'affinity'):
#    - 'euclidean': (Varsayılan) Kuş uçuşu mesafe (Pisagor).
#    - 'manhattan': Şehir bloğu mesafesi (L1 normu).
#    - 'cosine': Açısal benzerlik (Metin madenciliği vb. için).
#    - 'l1', 'l2': Diğer matematiksel normlar.
#
# 2. linkage (Bağlantı Kriteri - Kümeler arası mesafe nasıl ölçülür?):
#    - 'ward': (Varsayılan) Varyansı minimize eder (K-Means benzeri, dengeli kümeler).
#    - 'complete': En uzak noktalar arası mesafeyi kullanır (Aykırı değerlere hassas).
#    - 'average': Ortalama mesafeyi kullanır.
#    - 'single': En yakın noktalar arası mesafeyi kullanır (Zincirleme etkisi yaratabilir).
# -------------------------------------------------------------------------
hc = AgglomerativeClustering(n_clusters=optimum_k, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X_train)

print("✅ Kümeleme tamamlandı.")
print(f"📊 İlk 10 Verinin Kümesi: {y_hc[:10]}")
````

````python
# 3. GÖRSELLEŞTİRME (2D)
# ---------------------------------------------------
# Dinamik olarak tüm kümeleri çizdirir (Sizin kodunuzdaki sabit 3 renk yerine)
if hasattr(X_train, 'shape') and X_train.shape[1] >= 2:
    plt.figure(figsize=(12, 7))
    
    # Renk döngüsü (Otomatik)
    colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'orange', 'purple', 'brown']
    
    for i in range(optimum_k):
        color = colors[i % len(colors)] # Renk yetmezse başa dön
        plt.scatter(X_train[y_hc == i, 0], X_train[y_hc == i, 1], 
                    s=100, c=color, label=f'Küme {i+1}')

    plt.title(f'Hiyerarşik Kümeleme Sonuçları (K={optimum_k})')
    plt.xlabel('Öznitelik 1')
    plt.ylabel('Öznitelik 2')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("⚠️ Veri 2 boyuttan az, görselleştirme yapılamadı.")
````
