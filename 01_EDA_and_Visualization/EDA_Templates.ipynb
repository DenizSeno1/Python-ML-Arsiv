# 📊 Keşifsel Veri Analizi (EDA) Şablonları

Bu notebook, veriyi tanımak, aykırı değerleri görsel olarak hissetmek ve değişkenler arası ilişkileri okumak için sık kullandığın grafik şablonlarını içerir.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## 1. Tek Değişkenli Analiz (Univariate Analysis)

```python
def plot_distribution(df, column, bins=30):
    """
    Bir sütunun dağılımını (Histogram + KDE) ve kutu grafiğini (Boxplot) çizer.
    Aykırı değerleri ve çarpıklığı (skewness) tek bakışta görmek için idealdir.
    """
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    
    # Histogram ve KDE
    sns.histplot(df[column], bins=bins, kde=True, ax=ax[0], color="skyblue")
    ax[0].set_title(f'{column} Dağılımı', fontsize=14)
    
    # Boxplot (Aykırı değer tespiti için)
    sns.boxplot(x=df[column], ax=ax[1], color="lightgreen")
    ax[1].set_title(f'{column} Kutu Grafiği (Aykırı Değerler)', fontsize=14)
    
    plt.tight_layout()
    plt.show()

# Kullanım Örneği:
# plot_distribution(df_train, 'SalePrice')
```

## 2. Hedef Değişken ile İlişki (Bivariate Analysis)

```python
def plot_relation_with_target(df, feature_col, target_col, feature_type='num'):
    """
    Bir özelliğin hedef değişken ile ilişkisini çizer.
    feature_type: 'num' (sürekli sayısal veriler için scatterplot)
                  'cat' (kategorik veriler için boxplot)
    """
    plt.figure(figsize=(10, 6))
    
    if feature_type == 'num':
        sns.scatterplot(x=df[feature_col], y=df[target_col], alpha=0.6, color="coral")
        # Eğilim çizgisi eklemek istersen (opsiyonel)
        sns.regplot(x=df[feature_col], y=df[target_col], scatter=False, color="darkred")
        
    elif feature_type == 'cat':
        sns.boxplot(x=df[feature_col], y=df[target_col], palette="Set2")
        plt.xticks(rotation=45)
        
    plt.title(f'{feature_col} vs {target_col}', fontsize=15)
    plt.tight_layout()
    plt.show()

# Kullanım Örneği:
# plot_relation_with_target(df_train, 'GrLivArea', 'SalePrice', feature_type='num')
# plot_relation_with_target(df_train, 'Neighborhood', 'SalePrice', feature_type='cat')
```

## 3. Çoklu Değişken ve Korelasyon Matrisi

```python
def plot_correlation_matrix(df, top_k=None, target_col=None):
    """
    Verisetindeki sayısal sütunların korelasyon matrisini çizer (Multicollinearity tespiti için).
    Eğer top_k ve target_col verilirse, hedefe en çok etki eden K sütunu çizer.
    """
    # Sadece sayısal sütunları al
    num_df = df.select_dtypes(include=[np.number])
    
    if top_k and target_col:
        # Hedefle en çok korele olan K sütun
        corrmat = num_df.corr()
        cols = corrmat.nlargest(top_k, target_col)[target_col].index
        cm = np.corrcoef(num_df[cols].values.T)
        
        plt.figure(figsize=(10, 8))
        sns.set(font_scale=1.25)
        hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, 
                         yticklabels=cols.values, xticklabels=cols.values, cmap="coolwarm")
        plt.title(f'Hedef ile En Yüksek Korelasyona Sahip {top_k} Özellik', fontsize=16)
        plt.show()
    else:
        # Tüm sayısal sütunlar
        plt.figure(figsize=(14, 12))
        sns.heatmap(num_df.corr(), cmap="coolwarm", annot=False, vmin=-1, vmax=1)
        plt.title('Tüm Sayısal Değişkenlerin Korelasyon Matrisi', fontsize=16)
        plt.show()

# Kullanım Örneği:
# plot_correlation_matrix(df_train, top_k=10, target_col='SalePrice')
```
