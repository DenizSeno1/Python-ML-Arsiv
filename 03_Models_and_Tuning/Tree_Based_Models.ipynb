# 🌲 Ağaç Tabanlı Modeller (Tree-Based Models)

Kaggle yarışmalarının vazgeçilmezi olan algoritmaların (XGBoost, LightGBM, CatBoost ve Random Forest) kullanım şablonları.

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor, CatBoostClassifier

from sklearn.metrics import mean_squared_error, accuracy_score
```

## 1. Random Forest (Eski ama Güvenilir)

```python
def train_random_forest(X_train, y_train, X_test, task='regression'):
    """
    task: 'regression' veya 'classification'
    """
    if task == 'regression':
        model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    else:
        model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
        
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    return model, preds
```

## 2. XGBoost (Extreme Gradient Boosting)

```python
def train_xgboost(X_train, y_train, X_test, task='regression'):
    if task == 'regression':
        model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.05, random_state=42, n_jobs=-1)
    else:
        model = xgb.XGBClassifier(n_estimators=500, learning_rate=0.05, random_state=42, n_jobs=-1, eval_metric='logloss')
        
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    return model, preds
```

## 3. LightGBM (Hızlı ve Etkili)
Çok büyük verilerde XGBoost'tan daha hızlıdır. Kategorik değişkenleri kendisi tanıyabilir (ayrıca belirtilirse).

```python
def train_lightgbm(X_train, y_train, X_test, task='regression'):
    if task == 'regression':
        model = lgb.LGBMRegressor(n_estimators=500, learning_rate=0.05, random_state=42, n_jobs=-1)
    else:
        model = lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, random_state=42, n_jobs=-1)
        
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    
    return model, preds
```

## 4. CatBoost (Kategorik Veri Canavarı)
One-Hot Encoding yapmadan, kategorik sütunların isimlerini parametre olarak vererek eğitebilirsin. Target Encoding'i arka planda kendisi çok başarılı yapar.

```python
def train_catboost(X_train, y_train, X_test, cat_features=None, task='regression'):
    """
    cat_features: Kategorik sütunların indeksleri veya isimleri dizisi. Örn: ['Neighborhood', 'Condition1']
    """
    if task == 'regression':
        model = CatBoostRegressor(iterations=500, learning_rate=0.05, random_seed=42, verbose=100)
    else:
        model = CatBoostClassifier(iterations=500, learning_rate=0.05, random_seed=42, verbose=100)
        
    if cat_features:
        model.fit(X_train, y_train, cat_features=cat_features)
    else:
        model.fit(X_train, y_train)
        
    preds = model.predict(X_test)
    
    return model, preds
```
