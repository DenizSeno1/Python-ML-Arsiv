# 🧹 BACKWARD ELIMINATION (GERİYE DOĞRU ELEME)

> **🎯 AMAÇ:** Model için gereksiz değişkenleri (P-value yüksek olanları) elemek.
> **📥 GİRDİ:** X (Bağımsız Değişkenler), y (Hedef Değişken)
> **📤 ÇIKTI:** En anlamlı değişkenlerin listesi

### ⏱️ NE ZAMAN KULLANILIR?
- Çok fazla değişken varsa
- Modelin başarısını artırmak için
- Hangi değişkenin etkisiz olduğunu görmek için

````python
import statsmodels.api as sm
import pandas as pd

def backward_elimination(data, target, significance_level = 0.05):
    """
    Geriye Doğru Eleme (Backward Elimination) algoritması.
    
    Parametreler:
    data (pd.DataFrame): Bağımsız değişkenler (X)
    target (pd.Series): Hedef değişken (y)
    significance_level (float): Anlamlılık düzeyi (varsayılan 0.05)
    
    Döndürür:
    list: Seçilen en anlamlı değişkenlerin isimleri
    """
    features = data.columns.tolist()
    while(len(features) > 0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
            print(f"❌ Çıkarıldı: {excluded_feature} (P-value: {max_p_value:.4f})")
        else:
            break
    
    print("\n✅ Eleme Tamamlandı!")
    print(f"🎯 Kalan Özellikler ({len(features)} adet): {features}")
    return features

# KULLANIM ÖRNEĞİ (Eğer X ve y dataframe ise)
# ---------------------------------------------------
# selected_features = backward_elimination(X, y)
# X_optimal = X[selected_features]
````
